```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'FellowshipID':[1001,1002,1003,1004],
                    'Firstname':['Ajay','Vijay','Guna','Yuva'],
                    'Skill':['Hiding','Gardening','Spells','Fireworks']})
df1
```

```python
df2= pd.DataFrame({'FellowshipID':[1001,1002,1006,1007,1008],
                   'Firstname':['Ajay','Vijay','abdul','akhil','arjun'],
                   'Age':[22,25,34,3245,3456]})
df2
```

## Merge two dataframe

```python
df1.merge(df2)
```

## Merge dataframe

using some rules

```python
df1.merge(df2,how = 'inner',on = 'FellowshipID')
```

```python
df1.merge(df2,how = 'inner',on = ['FellowshipID','Firstname'])
```

# Outer join

```python
df1.merge(df2,how = 'outer')
```

## Left join

```python
df1.merge(df2,how = 'left')
```

## Right join

```python
df1.merge(df2,how = 'right')
```

## Cross join

```python
df1.merge(df2,how = 'cross')
```

## Required column

```python
df1[['Firstname','FellowshipID']].merge(df2[['Age','FellowshipID']],how = 'inner',on = 'FellowshipID')
```

# Join using join function

```python
df1.join(df2,how = 'outer',on = 'FellowshipID',lsuffix = '_left',rsuffix='_Right')
```

# Concatenation

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],join = 'inner')
```

```python
pd.concat([df1,df2],join = 'outer',axis = 1)
```

```python
df1._append(df2)
```
