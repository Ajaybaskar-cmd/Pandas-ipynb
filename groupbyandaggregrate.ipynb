```python
import pandas as pd
```

```python
df = pd.DataFrame({'Flavour':['Mini chococlate chip','chocolate','vannila','cookie dough','rockey road','pitachio','cake batter','neapoliten','brownie'],
                   'Base flavour':['vannila','chocolate','vannila','vannila','chocolate','vannila','vannila','vannila','chocolate'],
                   'Liked':['Yes','Yes','No','Yes','Yes','No','Yes','No','Yes'],
                   'Flavour rating':[10.0,8.8,4.7,6.9,8.2,2.3,6.5,3.8,8.2],
                   'Texture rating':[8.0,7.6,5.0,6.5,7.0,3.4,6.0,5.0,7.1],
                   'Total rating':[18.0,16.6,9.7,13.4,15.2,5.7,12.5,8.8,15.3]})
df
```

## First group by

```python
round(df.groupby('Base flavour').mean(numeric_only = True),2)
```

## Count of value
count is contains to string also

```python
df.groupby('Base flavour').count()
```

## Minimum

```python
df.groupby('Base flavour').min()
```

```python
df.groupby('Base flavour').max()
```

## Sum of value

```python
df.groupby('Base flavour').sum(numeric_only = True)
```

## Aggregrate function

One column using

```python
df.groupby('Base flavour').agg({'Flavour rating':['min','max','sum','mean']})
```

multiple column using

```python
df.groupby('Base flavour').agg({'Flavour rating':['min','max','sum','mean'],'Texture rating':['min','max','sum','mean']})
```

## Multiple column using group by

```python
df
```

```python
df.groupby(['Base flavour','Liked']).mean(numeric_only = True)
```

```python
df.groupby(['Base flavour','Flavour']).mean(numeric_only = True)
```

```python
df.groupby(['Base flavour','Liked']).agg({'Flavour rating':['mean','sum']})
```

## Describe using group by

```python
df.groupby('Base flavour').describe()
```
